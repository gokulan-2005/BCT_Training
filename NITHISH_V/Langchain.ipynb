```python
from transformers import pipeline
```

```python
chatbot = pipeline("text-generation", model="microsoft/DialoGPT-medium")
```

```python
memory = {}
```

```python
memory = {
    "user_name": None,
    "likes": None
}

```

```python
while True:
    user_input = input("You: ")

    if user_input.lower() == "exit":
        break

    text = user_input.lower()

    # Store name
    if "my name is" in text:
        name = user_input.lower().replace("my name is", "").strip().title()
        memory["user_name"] = name
        print(f"Bot: Nice to meet you, {name}!")
        continue

    # Store likes
    elif "i like" in text:
        like = user_input.lower().replace("i like", "").strip()
        memory["likes"] = like
        print(f"Bot: Got it! You like {like}.")
        continue

    # Answer name
    elif "what is my name" in text:
        if memory["user_name"]:
            print(f"Bot: Your name is {memory['user_name']}.")
        else:
            print("Bot: I don't know your name yet.")
        continue

    # Answer likes
    elif "what i like" in text or "what do i like" in text:
        if memory["likes"]:
            print(f"Bot: You like {memory['likes']}.")
        else:
            print("Bot: I don't know what you like yet.")
        continue

    # 🔹 Normal chat
    response = chatbot(
        user_input,
        max_new_tokens=50,
        temperature=0.7,
        do_sample=True,
        return_full_text=False
    )

    print("Bot:", response[0]["generated_text"])
```

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

model_name = "google/flan-t5-base"
```

```python
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)
```

```python
def generate_email(request):
    prompt = f"""
Write a professional email.

Request: {request}

Strict format:
Subject: <one line>

Dear Sir/Madam,

<Write a clear and polite message in 3-4 lines>

Thank you,
<Your Name>

Rules:
- Do not repeat sentences
- Keep it short and formal
"""

    inputs = tokenizer(prompt, return_tensors="pt", truncation=True)

    outputs = model.generate(
        **inputs,
        max_new_tokens=120,
        repetition_penalty=1.5,
        no_repeat_ngram_size=3
    )

    return tokenizer.decode(outputs[0], skip_special_tokens=True)
```

```python
# Loop
while True:
    user_input = input("Enter your request: ")

    if user_input.lower() == "exit":
        break

    print("\nGenerated Email:\n")
    print(generate_email(user_input))
    print("-" * 50)
```

```python
import os
os.environ["GROQ_API_KEY"] = ""
```

```python
from crewai import Agent, Task, Crew
```

```python
llm_model = "groq/llama-3.1-8b-instant"
```

```python
research_agent = Agent(
    role="Researcher",
    goal="Explain topic simply",
    backstory="You are a helpful teacher",
    llm=llm_model,
    verbose=True
)
```

```python
summary_agent = Agent(
    role="Summarizer",
    goal="Give short summary",
    backstory="You give bullet points",
    llm=llm_model,
    verbose=True
)
```

```python
research_task = Task(
    description="Explain Artificial Intelligence in simple terms",
    expected_output="Simple explanation",
    agent=research_agent
)
```

```python
summary_task = Task(
    description="Summarize into 5 bullet points",
    expected_output="5 bullet points",
    agent=summary_agent
)
```

```python
crew = Crew(
    agents=[research_agent, summary_agent],
    tasks=[research_task, summary_task],
    verbose=True
)
```

```python
result = crew.kickoff()
print(result)
```

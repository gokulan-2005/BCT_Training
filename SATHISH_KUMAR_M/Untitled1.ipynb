```python
from langgraph.graph import START, END, StateGraph
from typing import TypedDict
```

```python
class journeystate(TypedDict):
  curr_location: str
```

```python
def tuty(state: journeystate) -> journeystate:
  print("start the journey")
  return {"curr_location" :"Depart from tuty"}
```

```python
def madurai(state : journeystate) -> journeystate:
  print("Middle of the journey")
  return {"curr_location" :"middle from madurai"}
```

```python
def chennai(state : journeystate) -> journeystate:
  print("end of the journey")
  return {"curr_location" :"end from chennai"}
```

```python
builder = StateGraph(journeystate)
```

```python
builder.add_node("stt",tuty)
```

```python
builder.add_node("midd",madurai)
```

```python
builder.add_node("endd",chennai)
```

```python
builder.add_edge(START, "stt")
builder.add_edge("stt", "midd")
builder.add_edge("midd", "endd")
builder.add_edge("endd", END)
```

```python
graph = builder.compile()
```

```python
graph.invoke({"curr_location":"tuty"})
```
